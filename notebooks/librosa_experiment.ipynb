```python
import os
import pandas as pd
import librosa

import numpy as np
import matplotlib.pyplot as plt

os.chdir('../')
```

```python
file_path = "data//external//processed//"
meta_data = pd.read_csv("data//meta_data.csv")
```

```python
file_path =  os.path.join( file_path , meta_data.iloc[ 24 , 0 ] ) 
```

```python
sound , sr = librosa.load( file_path , sr=44100 )
```

```python
len(sound) == sr * 8 ## since our clip is 8 seconds long
```

```python
import numpy as np
import matplotlib.pyplot as plt

n_fft = 2048
D = np.abs(librosa.stft(sound[:n_fft], n_fft=n_fft, hop_length=n_fft+1))
plt.plot(D);
```

```python
D[0] , D[1]
```

```python
np.rfft( sound )
```

```python
from scipy.fft import fft, fftfreq ,rfft , rfftfreq

# Number of samples in normalized_tone
N = sr * 8

yf = rfft(sound)
xf = rfftfreq(N, 1 / sr)

plt.plot(xf, np.abs(yf))
plt.show()
```
